```python
lol_watcher = LolWatcher('RGAPI-d74ed218-6659-457b-933f-0e25b4c0b346')

my_region = 'kr'
```

```python
me = lol_watcher.summoner.by_name(my_region, '내앞에어둠이보여')

print(me)
```

```python
my_ranked_stats = lol_watcher.league.by_summoner(my_region, me['id'])
print(my_ranked_stats)
```
